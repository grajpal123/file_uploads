```python
#Creating plotting function

def myplot(data, playerlist=Players):
    
    Col = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue",
           "DwightHoward":"Magenta","ChrisBosh":"Dimgray","ChrisPaul":"Yellow","KevinDurant":"Lime",
           "DerrickRose":"Lavender","DwayneWade":"Magenta"}
    
    plt.rcParams['figure.figsize'] = 10,5
    
    for name in playerlist:
        plt.plot(data[Pdict[name]], color=Col[name], linestyle='--', marker='s', label=Players[Pdict[name]])
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation='vertical')
    plt.show()
```
