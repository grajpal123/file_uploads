## CONVERTING IMAGE - DIFFERENT FORMAT

    Here we will see how we can convert one format of an image be it RGB / RGBA into another format GRAY SCALE.

### `IMPORT LIBRARIES`

```python
import skimage
from skimage.io import imshow, imread, imsave
from skimage.color import rgb2gray, rgba2rgb
import numpy as np

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline 
```

```python
print('Skimage Version:', skimage.__version__)
```

<br>

### `LOADING IMAGE `

```python
# Loading RGB Image 

image_rgb = imread('dog.jpg')
imshow(image_rgb)
```

```python
# Loading RGBA Image 

image_rgba = imread('mid_alpha.png')
imshow(image_rgba)
```

<br>

### `CONVERTING IMAGES`

```python
# Converting RGBA -> RGB

image_rgba = imread('mid_alpha.png')
print("Shape of image before : {}".format(image_rgba.shape))

rgba_rgb_image = rgba2rgb(image_rgba)
print("Shape of image after  : {}".format(rgba_rgb_image.shape))

imshow(rgba_rgb_image)
```

    1) Nos. of channels is 3 which tells us that the image is in RGB format.

    2) We will also notice that the color and intensity of the image remain the same.

```python
# Converting RGB -> GRAY SCALE

image_rgb = imread('dog.jpg')
print("Shape of image before : {}".format(image_rgb.shape))

rgb_gray_image = rgb2gray(image_rgb)
print("Shape of image after  : {}".format(rgb_gray_image.shape))

imshow(rgb_gray_image)
```

    1) Nos. of channels reduced from 3 to 1 which implies that it is a grey scale image.

    2) Original image has also been converted into black and white image.


__NOTE__: _Now before loading the set of images we don't know in which format the images would be. It could be any format so, we will convert it into a standard gray scale format which reduces the size of the matrix and proceed further with our model building process_.

```python
image_gray1 = imread('dog.jpg', as_gray = True)
print(image_gray1.shape)
imshow(image_gray1)
```
